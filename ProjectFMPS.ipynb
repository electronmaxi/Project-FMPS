```python
import numpy as np
import math
from scipy import signal
import matplotlib.pyplot as plt
```

```python
N = 2**10 + 1 # 1025
L = np.pi 
w0 = 2
t = np.linspace(-L, L, N)
w = np.copy(t) 
omega = np.copy(t) 
fig, ax = plt.subplots(1,1,figsize=(18,4))
o1 = ax.plot(t, t, 'b-',  label='$t$')
ax.legend()
```

1. $x(t)=2\sin(\omega_0t)$
$$\hat x(\omega) = \frac{2 \pi}{i}  \left( \delta(\omega-\omega_0) - \delta(\omega+\omega_0)   \right)  $$

y luego

$$|\hat x(\omega)| = 2\pi \left( \delta(\omega-\omega_0) + \delta(\omega+\omega_0)  \right) $$

```python
def grafandmodulw1(w0):
    x = 2*np.sin(w0*t)
    n_w0 = int((L+w0) * N/(2*L))
    minus_n_w0 = int((L-w0) * N/(2*L))
    xhatmod = 2*np.pi*(signal.unit_impulse(N, n_w0) + 2*0.5*signal.unit_impulse(N, minus_n_w0))
    fig, ax = plt.subplots(1, 2,figsize=(18,4))
    o1 = ax[0].plot(t, x, 'b-',  label='$x$')
    o2 = ax[1].plot(w, xhatmod, 'r-', label='modulo de $\hat x$')
    ax[0].legend()
    ax[1].legend()
```

### Ejercicio 2
Calcule la transformada de fourier.
Para las siguiente se침ales, grafique la se침al, calcule la transformada de fourier de las se침ales y grafique su modulo.

1.  Si $x(t) = e^{-at}\gamma(t)$, entonces



\begin{eqnarray}
  \hat x(\omega) &=& \int_0^\infty e^{-at} e^{-i \omega t} dt  \\
                 &=& \int_0^\infty e^{-(a +i \omega) t} dt \\
                 &=& \left[ \frac{e^{-(a +i \omega) t}}{-(a +i \omega)} \right]_0^\infty \\
                 &=&  \frac{1}{(a +i \omega)}
\end{eqnarray}  

```python
for a in [1, .1, .01]:
    
    x = np.exp(-a*t) 
    x[:int(N/2)-1] = 0
    xhat = 1/(a+1j*omega)
    fig, ax = plt.subplots(1,2,figsize=(18,4))
    o1 = ax[0].plot(t, x, 'b-', label='$x$')
    o2 = ax[1].plot(w, np.real(xhat), 'g-', label='parte real de $\hat x$')
    o2 = ax[1].plot(w, np.imag(xhat), 'm-', label='parte imaginaria de $\hat x$')
    o2 = ax[1].plot(w, np.abs(xhat), 'r-', label='modulo de $\hat x$')
    ax[0].legend()
    ax[1].legend()
    ax[0].set_title(f'$a = {a}$')
    ax[1].set_title(f'$a = {a}$')
```

2.  Si $x(t) = e^{-a|t|}$, entonces



\begin{eqnarray}
  \hat x(\omega) &=& \int_{-\infty}^0 e^{at} e^{-i \omega t} dt+\int_0^\infty e^{-a|t|} e^{-i \omega t} dt  , como t siempre es positivo es lo mismo\\
                 &=& -\int_{0}^{-\infty} e^{t(a-iw)} dt+\int_0^\infty e^{-(a +i \omega) t} dt \\
                 &=& -\left[ \frac{e^{(a -i \omega) t}}{(a -i \omega)} \right]_0^{-\infty} + \left[ \frac{e^{-(a +i \omega) t}}{-(a +i \omega)} \right]_0^\infty \\
                 &=&  \frac{1}{(a -i \omega)} +\frac{1}{(a +i \omega)} 
\end{eqnarray}  

```python
for a in [1, .1, .01]:
    x = np.exp(-a*np.abs(t))
    xhat = 2*a/(np.power(a,2)+np.power(omega,2))
    fig, ax = plt.subplots(1,2,figsize=(18,4))
    o1 = ax[0].plot(t, x, 'b-', label='$x$')
    o2 = ax[1].plot(w, np.real(xhat), 'g-', label='parte real de $\hat x$')
    o2 = ax[1].plot(w, np.imag(xhat), 'm-', label='parte imaginaria de $\hat x$')
    o2 = ax[1].plot(w, np.abs(xhat), 'r-', label='modulo de $\hat x$')
    ax[0].legend()
    ax[1].legend()
    ax[0].set_title(f'$a = {a}$')
    ax[1].set_title(f'$a = {a}$')
```

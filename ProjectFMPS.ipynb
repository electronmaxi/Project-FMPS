### Ejercicio 1
Transformada de Fourier y representación frecuencial.
Usando las propiedades de transformada de fourier calcule la transformada de las siguientes señales. Grafique la señal x y el modulo de la transformada.


Importación y dominios preliminares

```python
import numpy as np
import math
from scipy import signal
import matplotlib.pyplot as plt
```

```python
N = 2**10 + 1 # 1025
L = np.pi 
w0 = 2
t = np.linspace(-L, L, N)
w = np.copy(t) 
omega = np.copy(t) 
fig, ax = plt.subplots(1,1,figsize=(18,4))
o1 = ax.plot(t, t, 'b-',  label='$t$')
ax.legend()
```

1. $x(t)=2\sin(\omega_0t)$

```python
def grafandmodulw1(w0):
    x = 2*np.sin(w0*t)
    n_w0 = int((L+w0) * N/(2*L))
    minus_n_w0 = int((L-w0) * N/(2*L))
    xhatmod = 2*np.pi*(signal.unit_impulse(N, n_w0) + 2*0.5*signal.unit_impulse(N, minus_n_w0))
    fig, ax = plt.subplots(1, 2,figsize=(18,4))
    o1 = ax[0].plot(t, x, 'b-',  label='$x$')
    o2 = ax[1].plot(w, xhatmod, 'r-', label='modulo de $\hat x$')
    ax[0].legend()
    ax[1].legend()
```

Grafico de la señal x y el modulo de la transformada con $\omega_0=1$

```python
grafandmodulw1(1)
```

Grafico de la señal x y el modulo de la transformada con $\omega_0=-1$

```python
grafandmodulw1(-1)
```

Grafico de la señal x y el modulo de la transformada con $\omega_0=2$

```python
grafandmodulw1(2)
```

2. $x(t)=2\sin(2\omega_0t +\frac{2}{3})$

```python
def grafandmodulw2(w0):
    x = 2*np.sin(2*w0*t+2/3)
    n_w0 = int((L+2*w0) * N/(2*L))
    minus_n_w0 = int((L-2*w0) * N/(2*L))
    xhatmod = 2*np.pi*(signal.unit_impulse(N, n_w0) + 2*0.5*signal.unit_impulse(N, minus_n_w0))
    fig, ax = plt.subplots(1, 2,figsize=(18,4))
    o1 = ax[0].plot(t, x, 'b-',  label='$x$')
    o2 = ax[1].plot(w, xhatmod, 'r-', label='modulo de $\hat x$')
    ax[0].legend()
    ax[1].legend()
```

```python
grafandmodulw2(1)
```

3. $x(t)=2\cos(\omega_0t)+\sin(\omega_0t)$

```python
def grafandmodulw3(w0):
    x = 2*np.cos(w0*t)+np.sin(w0*t)
    n_w0 = int((L+2*w0) * N/(2*L))
    minus_n_w0 = int((L-2*w0) * N/(2*L))
    xhatmod = 2*np.pi*math.sqrt(5)*(signal.unit_impulse(N, n_w0) + 2*0.5*signal.unit_impulse(N, minus_n_w0))
    fig, ax = plt.subplots(1, 2,figsize=(18,4))
    o1 = ax[0].plot(t, x, 'b-',  label='$x$')
    o2 = ax[1].plot(w, xhatmod, 'r-', label='modulo de $\hat x$')
    ax[0].legend()
    ax[1].legend()
```

```python
grafandmodulw3(1)
```

```python
grafandmodulw3(-0.5)
```
